```python
#importing libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df1 = pd.read_csv('/content/iris_1.csv')
```

```python
df1.head()
```

```python
df1.shape
```

```python
#Performing Univariate Analysis 

df_setosa = df1.loc[df1['species']=='setosa'].head()
```

```python
df_setosa
```

```python

df_virginica = df1.loc[df1['species']=='virginica'].head()


df_versicolor = df1.loc[df1['species']=='versicolor'].head()

```

```python
df_versicolor
```

```python
plt.plot(df_setosa['petal_length'], np.zeros_like(df_setosa['petal_length']))
plt.plot(df_virginica['petal_length'], np.zeros_like(df_setosa['petal_length']))
plt.plot(df_versicolor['petal_length'], np.zeros_like(df_setosa['petal_length']))
plt.xlabel('Petal Lenght')
```

Np_Zeros is used to return 0 values on the Y axis

Insights
We can see how we are focussing on only one feature ie- Petal Lenght for all the species. 
X axis represents the Petal Lenght for all the species 
The colors blue, green and prange represent different species.

Bivariate Analysis 

```python
#Bivariate Analysis 

sns.FacetGrid(df1,hue = 'species', size = 6).map(plt.scatter, 'petal_length', 'sepal_width').add_legend()
plt.show()
```

From the above figure we can analyse that Bivariate Anlaysis has been performed. 
Two feautres have been used - Petal Lenght and Sepal Widht and they have been plotted WRT to their hue. 

From the above figure, we can see that the Setosa species has a linear relation but Versicor and Virginica are overlapping. 


Multivariate Analysis 


```python
#Multivariate Analysis 
sns.pairplot(df1, hue='species', size =2)
```

The above figure shows us the implementation of Multivariate Anlaysis. 
The usage of multivariate analysis is to comapre more than two feautres at the same time. 

This analysis, helps us to compare various coorelations between more than one graph.
